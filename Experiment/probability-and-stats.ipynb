### Binomial Distribution ~ Bin(n,p)
Bernoulli experiments lead to an important and more interesting model. The binomial distribution arises when a Bernoulli experiment is repeated n times.
Let's take random variable X = 1 when someone wins spin a wheel on cred and 0 when he loses

P(X=1) = p = $\pi$ =0.15

He has 10 spins with him

```python
X = np.arange(1, 11)
n, p = 10, 0.15
```

```python
mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')
```

```python
print(pd.DataFrame([
    ['Mean', mean],
    ['Variance', var],
    ['Skewness', skew],
    ['Kurtosis', kurt],
],
                   columns=['Statistics', 'Value'],)
                    )
```

```python
dist = binom(n,p)
```

```python
plt.bar(X, dist.pmf(X))
plt.xlabel('No. of spin prize won')
plt.ylabel('Probability of spin prize winning')
plt.title('PMF of spin prize winning for 10 chances')
```

```python
plt.bar(X, dist.cdf(X))
plt.xlabel('No. of spin prize won')
plt.ylabel('Probability of spin prize winning')
plt.title('PMF of spin prize winning for 10 chances')
```

Suppose you're randomly picking a point along a straight road that stretches from 0 to 100 meters. The location you pick is equally likely to be anywhere on that road. This is a uniform continuous distribution, where any point between 0 and 100 meters has the same probability of being selected. (Case of delivery partner with weak signal near location)
X ∼ U(0,100)

### Normal (Gaussian) Distribution ∼N(μ,$σ^{n}$)
The heights of people in a large city. Most people will have average heights close to the city’s average height, with fewer people being extremely tall or short. Heights tend to follow a normal distribution, with the majority of people clustered around the mean height and fewer people at the extremes.
The heights of people in a city, where the average height is 170 cm and the standard deviation is 10 cm:
X ~ N (170, $10^{2}$)

### Standard Normal (Gaussian) Distribution
Example: After standardizing the scores of students in an exam (subtracting the mean score and dividing by the standard deviation), you get a standard normal distribution, where the mean is 0 and the standard deviation is 1. This helps compare how far a student's score is from the average in terms of standard deviations.
After standardizing exam scores with a mean of 50 and standard deviation of 10, you get:
X∼N(0,1) (standard normal distribution).

### Exponential Distribution ∼Exp(λ)
Example: The time between incoming customer service calls at a busy store. The store may get calls randomly, but the time between calls is typically exponentially distributed. If the average wait time between calls is 10 minutes, you can predict how likely it is that the next call will come in the next 5 minutes, or 15 minutes, etc.

The time between incoming customer service calls with an average of 10 minutes (i.e., X∼Exp(0.1)

### Triangular Distribution ∼Triangular(a,b,c)
Example: When estimating how long it takes to complete a task, you might have a good idea of the minimum (10 minutes), most likely (15 minutes), and maximum (25 minutes) time it could take. The triangular distribution helps model this uncertainty, where the most likely time is the peak, and the distribution tapers off as you move toward the extremes.
X∼Triangular(10,25,15)
